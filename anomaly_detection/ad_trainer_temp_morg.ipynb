```python
# Train the autoencoder
!python /content/Jet-Image-Tagging/anomaly_detection/src/train.py \
    --data_path "/content/jet_images_128.h5" \
    --save_dir "/content/drive/MyDrive/JetTagging/anomaly_detection/checkpoints" \
    --epochs 30 \
    --latent_space_dim 128 \
    --batch_size 64 \
    --lr 0.0005 \
    --encoded_space_dim 128 \
    --bg_classes 0 1 \
    --img_size 128 \
    --weight_decay 0.0001 \
    --max_samples 50000 \
    --patience 10 \
    --resume_from "/content/drive/MyDrive/JetTagging/anomaly_detection/checkpoints/autoencoder_latest.pth"
```
