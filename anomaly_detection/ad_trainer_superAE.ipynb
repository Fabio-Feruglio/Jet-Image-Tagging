## 2. Autoencoder with denoising and signal reconstruction priority training

```python
#Training
!python /content/Jet-Image-Tagging/anomaly_detection/src/train_dae.py \
    --data_path "/content/jet_images_128.h5" \
    --save_dir "/content/drive/MyDrive/JetTagging/anomaly_detection/checkpoints/dae" \
    --epochs 40 \
    --latent_space_dim 128 \
    --batch_size 128 \
    --lr 0.0005 \
    --active_weight 5.0 \
    --noise_factor 0.2 \
    --bg_classes 0 1 \
    --img_size 128 \
    --weight_decay 0.0001 \
    --max_samples 50000 \
    --patience 10 \
    --resume_from "/content/drive/MyDrive/JetTagging/anomaly_detection/checkpoints/dae/dae_latest.pth"
```
