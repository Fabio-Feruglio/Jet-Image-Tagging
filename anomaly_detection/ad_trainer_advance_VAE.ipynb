## 2. Autoencoder training

```python
# Lancia l'addestramento
# Nota i nuovi parametri: latent_space_dim=32, target_beta=0.1, warmup_epochs=10
!python /content/Jet-Image-Tagging/anomaly_detection/src/train_advanced_vae.py \
    --data_path "/content/jet_images_128.h5" \
    --save_dir "/content/drive/MyDrive/JetTagging/anomaly_detection/checkpoints/vae_advanced" \
    --epochs 40 \
    --latent_space_dim 32 \
    --batch_size 64 \
    --lr 0.0005 \
    --target_beta 0.1 \
    --warmup_epochs 10 \
    --bg_classes 0 1 \
    --img_size 128 \
    --weight_decay 0.0001 \
    --max_samples 50000
```

```python
# Lancia la valutazione
# Genererà un grafico ROC con tre curve: MSE, KL e Isolation Forest
!python /content/Jet-Image-Tagging/anomaly_detection/src/evaluation_advanced_vae.py \
    --model_path "/content/drive/MyDrive/JetTagging/anomaly_detection/checkpoints/vae_advanced/vae_advanced_best.pth" \
    --max_samples 50000 \
    --latent_space_dim 32 \
    --batch_size 64 \
    --img_size 128 \
    --bg_classes 0 1 \
    --data_path "/content/jet_images_128.h5" \
    --save_dir "/content/drive/MyDrive/JetTagging/anomaly_detection/results/vae_advanced"
```
