## Launch the pre-processing function
The pre-processing function downloads the JetNet dataset (if it has not been done already) and converts the information on the particle content in each jet into an image, using the native to_image method offered by the JetNet package. The output image dataset is saved in a compressed .h5 file. 
Args for the dataset_preprocessing.py file:
- --output_file : Output file path name (should be a .h5 file)
- --data_dir    : Raw jetnet dataset directory
- --batch_size  : Jet batch size for vectorized conversion and saving
- --im_size     : Output images size in pixel (im_size x im_size)
- --maxR        : Maximum range for pseudorapidity and azimuthal angle coordinates
- --compression : Type of compression method (gzip or lzf) (jet images are highly sparse)
- --compression_opts    : Gzip compression level
- --dtype       : Precision for image saving (float16 or float32)
